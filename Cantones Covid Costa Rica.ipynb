```python
import pandas as pd
```

```python
activos = pd.read_csv('datos/07_29_CSV_ACTIVOS.csv', encoding='iso-8859-1')
activos.head([])
```

```python
activos_alajuela = activos[activos["provincia"]=='Alajuela']
activos_alajuela
```

```python
def filterandplot(ds, codCanton):
    activos_canton = ds[ds["cod_canton"]==codCanton]
    transposed = activos_canton.melt(id_vars=activos_canton.columns[0:4])
    transposed.head()
    transposed.plot.line(x="variable", y="value")
    return activos_canton
    
```

```python
cod_poas = 208
cod_grecia = 203
activos_poas= filterandplot(activos_alajuela, cod_poas)
activos_grecia=filterandplot(activos_alajuela, cod_grecia)
```
